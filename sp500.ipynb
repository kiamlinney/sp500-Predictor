```python
def predict(train, test, predictors, model):
    model.fit(train[predictors], train["Target"])
    predictions = model.predict(test[predictors])
    predictions = pd.Series(predictions, index=test.index, name="Predictions")
    combined = pd.concat([test["Target"], predictions], axis=1)
    return combined
```

```python
def backtest(data, model, predictors, start=2500, step=250): #take 10 years of data and train first model with 10 years of data, step is 250 because model is trained for about a year
    #taking the first 10 years of data to predict values for the 11th year, then taking the first 11 years of data to predict values for the 12 year, and so on.
    all_predictions = []

    for i in range(start, data.shape[0], step):
        train = data.iloc[0:i].copy()
        test = data.iloc[i: (i + step)].copy()
        predictions = predict(train, test, predictors, model)
        all_predictions.append(predictions)

    return pd.concat(all_predictions)
```

```python
predictions = backtest(sp500, model, predictors)
```

```python
horizons = [2, 5, 60, 250, 1000] # Will calculate mean closing price in last 2 days, last week, last 3 months, last year, and last 4 years, then will find the ratio between today's closing price and the price in those periods
new_predictors = []

for horizon in horizons:
    rolling_avgs = sp500.rolling(horizon).mean()
    ratio_col = f"Close_Ratio_{horizon}"
    sp500[ratio_col] = sp500["Close"] / rolling_avgs["Close"]

    trend_col = f"Trend_{horizon}"
    sp500[trend_col] = sp500.shift(1).rolling(horizon).sum()["Target"]

    new_predictors += [ratio_col, trend_col]
    # On any given day, will look at the past few days, and find the sum of the target and compute a rolling sum.
```

```python
sp500 = sp500.dropna() # Drop all the NaN columns
```

```python
# Now to improve the model:
```

```python
model = RandomForestClassifier(n_estimators = 200, min_samples_split = 50, random_state = 1)
```

```python
def predict(train, test, predictors, model):
    model.fit(train[predictors], train["Target"])
    predictions = model.predict_proba(test[predictors])[:,1] # Will now return the probability that the row will be a 0 or a 1, get second column, probability that the price goes up
    predictions[predictions >= .6] = 1
    predictions[predictions < .6] = 0
    # Setting custom threshold, forcing model to be more confident that the price will go up. Default threshold is .5
    predictions = pd.Series(predictions, index=test.index, name="Predictions")
    combined = pd.concat([test["Target"], predictions], axis=1)
    return combined

    # Will reduce the total number of trading days, but will increase the chance that the price will actually go up on those days
```

```python
predictions = backtest(sp500, model, new_predictors)
```

```python
predictions["Predictions"].value_counts()
```

```python
# Now distribution between days predicted to go up and predicted to go down, there are fewer predicted to go up due to new threshold.
```

```python
precision_score(predictions["Target"], predictions["Predictions"])
```

```python
# When the model predicts that the price will go up, approx. 57% of the time it will actually go up.
```
